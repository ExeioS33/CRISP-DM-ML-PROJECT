# Feature Selection ✅

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split

# Drop unnecessary index column if it exists
if "Unnamed: 0" in merged_data.columns:
    merged_data.drop(columns=["Unnamed: 0"], inplace=True)

# 🔹 **Step 2: Encode Categorical Variables**
merged_data["sex"] = merged_data["sex"].map({"Male": 1, "Female": 0})

physical_activity_mapping = {"Low": 0, "Intermediate": 1, "High": 2, "Average": 3}
merged_data["physical_activity"] = merged_data["physical_activity"].map(physical_activity_mapping)

merged_data["target"] = merged_data["target"].map({"heart_disease": 1, "normal": 0})

# Ensure all data is numeric before feature selection
numeric_data = merged_data.select_dtypes(include=["int64", "float64"])
```

```python
# 🔹 **Step 3: Apply VarianceThreshold for Feature Selection**
# Exclude 'target' before feature selection
X = numeric_data.drop(columns=["target"])

# Apply VarianceThreshold
selector = VarianceThreshold(threshold=0.01)
selector.fit(X)

# Select features with variance above the threshold
selected_features = X.loc[:, selector.get_support()]

# Add back the target column
selected_features["target"] = numeric_data["target"]
```

#### **1️⃣ ST Slope (`0.51`)**  
   - Strongest positive correlation.  
   - **Higher ST slope values** (often “upsloping” in ECG readings) are frequently associated with heart disease.  

#### **2️⃣ Exercise-Induced Angina (`0.48`)**  
   - Patients experiencing **chest pain (angina) during exercise** are more likely to have heart disease.  

#### **3️⃣ Chest Pain Type (`0.46`)**  
   - Different **chest pain types** moderately correlate with disease presence.  
   - Higher "type" categories (like "typical" or "non-anginal") are often associated with higher risk.  

#### **4️⃣ ST Depression (Oldpeak) (`0.40`)**  
   - Higher **ST depression values** (induced by exercise) are linked to heart disease.  

#### **5️⃣ Maximum Heart Rate (`-0.41`)**  
   - **Negative correlation**:  
     - Patients with **higher max heart rates** during exercise **tend to have a lower incidence** of heart disease.  

#### **6️⃣ Sex (`0.31`)**  
   - **Slightly positive correlation**, indicating that **males (if sex=1) may have a higher likelihood** of heart disease.
